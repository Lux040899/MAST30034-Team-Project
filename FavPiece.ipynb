```python
dfScsful = df.loc[((df['WElo']>2000) & (df['WIsComp'].isnull())) | ((df['BElo']>2000) & (df['BIsComp'].isnull()))].reset_index(drop=True)
dfScsful.shape
```

```python
def formatMoves(moves):
    formattedMoves = []
    moves=moves.strip("[]").split(',')
    for i in moves:
        i = i.strip()
        i = i.strip("'")
        formattedMoves.append(i)
    return formattedMoves
```

```python
formatMoves(dfScsful.loc[2, 'WMoves'])
```

```python
WPieceMoves = {'N': 0, 'B': 0, 'R': 0, 'Q': 0, 'K':0}
BPieceMoves = {'N': 0, 'B': 0, 'R': 0, 'Q': 0, 'K':0}
```

```python
dfScsful['WFavPiece'] = 'null'
dfScsful['BFavPiece'] = 'null'
```

```python
for i in range(len(dfScsful)):
    if (dfScsful.loc[i]['WElo'] >= 2000):
        moves = formatMoves(dfScsful.loc[i, 'WMoves'])
        for move in moves:
            if(move):
                if move[0].isupper() and move[0] != 'O':
                    WPieceMoves[move[0]] += 1
        FavPiece = max(WPieceMoves, key=WPieceMoves.get)
        dfScsful.loc[i, ['WFavPiece']] = FavPiece
        
    
    elif (dfScsful.loc[i]['BElo'] >= 2000):
        moves = formatMoves(dfScsful.loc[i, 'WMoves'])
        for move in moves:
            if(move):
                if move[0].isupper() and move[0] != 'O':
                    BPieceMoves[move[0]] += 1
        FavPiece = max(BPieceMoves, key=BPieceMoves.get)
        dfScsful.loc[i, ['BFavPiece']] = FavPiece
```
