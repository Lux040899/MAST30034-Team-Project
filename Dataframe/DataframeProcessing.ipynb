```python
f = open("data.pgn")
line = f.readline()
while line !=None:
    row = []
    row.append(line[8:-3])
    line = f.readline()
    row.append(line[7:-3])
    line = f.readline()
    row.append(line[20:-3])
    line = f.readline()
    row.append(line[8:-3])
    line = f.readline()
    row.append(line[8:-3])
    line = f.readline()
    row.append(line[11:-3])
    line = f.readline()
    row.append(line[11:-3])
    line = f.readline()
    row.append(line[10:-3])
    line = f.readline()
    row.append(line[10:-3])
    line = f.readline()
    if line[1] == "W":
        row.append(line[14:-3])
        line = f.readline()
    else:
        row.append(None)
    if line[1] == "B":
        row.append(line[14:-3])
        line = f.readline()
    else:
        row.append(None)
    row.append(line[14:-3])
    line = f.readline()
    row.append(line[7:-3])
    line = f.readline()
    row.append(line[7:-3])
    line = f.readline()
    row.append(line[13:-3])
    line = f.readline()
    row.append(line[13:-3])
    line = f.readline()
    row.append(line[6:-3])
    line = f.readline()
    row.append(line[11:-3])
    line = f.readline()
    row.append(line[9:-3])
    line = f.readline()
    line = f.readline()
    wmoves = []
    bmoves = []
    j = 0
    
    while True:
        if line[j] == "{":
            break
        while line[j] != " ":
            j += 1
        j += 1
        if line[j] == "{":
            break
        move = ""
        while line[j] != " ":
            move += line[j]
            j += 1  
        wmoves.append(move)
        j += 1
        move = ""
        if line[j] == '{':
            break
        while line[j] != " ":
            move += line[j]
            j += 1
        bmoves.append(move)
        j += 1
        if j >= len(line):
            break
    row.append(wmoves)
    row.append(bmoves)
    f.readline()
    line = f.readline()
    df = df.append(pd.Series(row, index = df.columns), ignore_index = True)
    
```

```python
df.to_csv("data.csv")
```
