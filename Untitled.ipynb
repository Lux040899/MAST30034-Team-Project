```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("sample.csv")
df.drop('Unnamed: 0', axis=1, inplace = True)
```

```python
df['WNumofMoves'] = 0
df['BNumofMoves'] = 0
```

```python
def formatMoves(moves):
    formattedMoves = []
    moves=moves.strip("[]").split(',')
    for i in moves:
        i = i.strip()
        i = i.strip("'")
        formattedMoves.append(i)
    return formattedMoves
```

```python
for i in range(len(df)):
    WMoves = formatMoves(df.loc[i, 'WMoves'])
    BMoves = formatMoves(df.loc[i, 'BMoves'])
    moveCount = 0
    for move in WMoves:
        if(move):
            moveCount += 1  
    df.loc[i, ['WNumofMoves']] = moveCount
    moveCount = 0
    for move in BMoves:
        if(move):
            moveCount += 1
    df.loc[i, ['BNumofMoves']] = moveCount
```
