```python
import pandas as pd
import ast
from collections import defaultdict
```

```python
df = pd.read_csv("sample.csv")
```

```python
df = pd.read_csv("sample.csv")
df["BMoves"] = df["BMoves"].apply(ast.literal_eval)
df["WMoves"] = df["WMoves"].apply(ast.literal_eval)
df["BElo"] = df["BElo"].astype(int)
df["WElo"] = df["WElo"].astype(int)
```

```python
def find_king(moves, start):
    for i in range (-1, -(len(moves) +1), -1):
        if moves[i][0] == "K":
            return moves[i][-2:]
        elif moves[i] == "O-O":
            if (start == "e1"):
                return "g1"
            else:
                return "g8"
        elif moves[i] == "O-O-O":
            if (start == "e1"):
                return "c1"
            else:
                return "c8"
    return start
```

```python
def common_death_squares(df):
    dc = defaultdict(int)
    for i in range(len(df)):
        entry = df.iloc[i]
        if entry["Result"] == "1-0":
            dc[find_king(entry["BMoves"], "e8")] += 1
        elif entry["Result"] == "0-1":
            dc[find_king(entry["WMoves"], "e1")] += 1
    return {k: v for k, v in sorted(dc.items(), key=lambda item: item[1])}
            
```

```python
dc = common_death_squares(df)
```

```python
def high_skill_openings(df):
    open_elo = defaultdict(int)
    open_count = defaultdict(int)
    for i in range(len(df)):
        wOpen = tuple(df["WMoves"][i][0:5])
        wSkill = df["WElo"][i]
        open_elo[wOpen] += wSkill
        open_count[wOpen] += 1
        bOpen = tuple(df["BMoves"][i][0:5])
        bSkill = df["BElo"][i]
        open_elo[bOpen] += bSkill
        open_count[bOpen] += 1
        
    for key in open_elo.keys():
        open_elo[key] = open_elo[key]/open_count[key]
    return open_elo
    
```

```python
high_skill_openings(df)
```
